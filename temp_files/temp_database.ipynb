```python
import pandas as pd
import numpy as np
import json
```

```python
df = pd.DataFrame(np.nan, index = range(0,1000), columns = ['Name','Subject','Grade', 'District', 'Type', 'Tags'])

names = pd.read_csv('./names.csv')
names_list = names['name'].tolist()

BC_district = pd.read_csv('./BC_school_districts.csv')
ON_district = pd.read_csv('./ON_school_districts.csv')
BC_district_list = BC_district['district'].tolist()
ON_district_list = ON_district['district'].tolist()
district_list = BC_district_list + ON_district_list

grade = pd.read_csv('./grade.csv')
grade_list = grade['grade'].tolist()

subject = pd.read_csv('./subject.csv')
subject_list = subject['subject'].tolist()

tags = pd.read_csv('./tags.csv')
tags_list = tags['tag'].tolist()

type = pd.read_csv('./type.csv')
type_list = type['type'].tolist()

grade = pd.read_csv('./grade.csv')
grade_list = grade['grade'].tolist()
```

```python
df.head()

df['Name1'] = np.random.choice(names_list, len(df))
df['Name2'] = np.random.choice(names_list, len(df))
df['Name3'] = np.random.choice(names_list, len(df))
df['Name'] = df['Name1'] + ' ' + df['Name2'] + ' ' + df['Name3']
df = df.drop(['Name1', 'Name2', 'Name3'], axis=1)

df['Subject'] = np.random.choice(subject_list, len(df))
df['Grade'] = np.random.choice(grade_list, len(df))
df['District'] = np.random.choice(district_list, len(df))
df['Type'] = np.random.choice(type_list, len(df))
df['Tags'] = np.random.choice(tags_list, len(df))

df
```
